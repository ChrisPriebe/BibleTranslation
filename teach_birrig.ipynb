```python
import pandas as pd
import lib.cipher as cipher

bible = pd.read_csv('data/bible.csv')
bible = bible[bible['book'].isin(['MAT','LUK','JHN'])]
bible
```

```python
# count all the words in the column birrig and put the in a dictionary
# where the key is the word and the value is the count
birrig_counts = {}
for text in bible['birrig']:
    if type(text) != str:
        continue
    for word in text.split():
        # strip from the word anything not a-zA-Z
        word = ''.join([c for c in word if c.isalpha()]).lower()
        if word == '':
            continue
        if word in birrig_counts:
            birrig_counts[word] += 1
        else:
            birrig_counts[word] = 1

# for each word in the dictionary, encrypt it and put it in a new dictionary
# where the key is the encrypted word and the value is the count
translation_index = {}
for word, count in birrig_counts.items():
    if count < 50:
        continue
    encrypted_word = cipher.substitution_cipher(word, encode=False)
    translation_index[word] = encrypted_word
len(translation_index)
```
