# Fetch the Bible in various translations and prepare it

The Bible are available at https://github.com/BibleNLP and they are aligned verse by verse
seperated by newlines.  The reference for them are in vref.txt and each version is in their own
file.  We can download them and put them into a pandas dataframe

````python
import requests
import pandas as pd

def download_references():
  
    url = 'https://github.com/BibleNLP/ebible/blob/main/metadata/vref.txt?raw=true'
    r = requests.get(url)
    if r.status_code != 200:
        raise Exception(f'Failed to download {url}')

    # create a dataframe and read each line into a row
    df = pd.DataFrame(r.text.splitlines())
    
    # for each row extract the first 3 characters as book
    df['book'] = df[0].str[:3]
    # for each row extract the 4th character up till : as chapter
    df['chapter'] = df[0].str[3:].str.split(':').str[0]
    # and the remainder as verse
    df['verse'] = df[0].str[3:].str.split(':').str[1]

    return df


def clean_filename(filename):
    """ Security function to ensure filename is a valid filename of a-zA-Z0-9_- """
    return ''.join([c for c in filename if c.isalnum() or c in '-_'])

# download bible from https://github.com/BibleNLP/ebible/blob/main/corpus/eng-eng-web.txt?raw=true
# and save it as bible.txt
def download_bible_to_dataframe(language='eng', version='eng-web', df=None):
    """
    Download a bible version from BibleNLP and save it to a dataframe
    :param language: language code (3 char code - check filename on BibleNLP
    :param version: version code (the remaining name of the file - check filename on BibleNLP)
    :param df: optional dataframe to append to (if empty will initialize a new dataframe)
    :return: dataframe with the bible version added

    """
    if df is None:
        df = download_references()

    # for security reasons ensure versions is a valid filename of a-zA-Z0-9_-
    version = clean_filename(version)
    language = clean_filename(language)
    
    url = f'https://github.com/BibleNLP/ebible/blob/main/corpus/{language}-{version}.txt?raw=true'
    r = requests.get(url)
    if r.status_code != 200:
        raise Exception(f'Failed to download {url}')
    df[version] = r.text.splitlines()
    return df


````

## Why use encoding
To get around this "cheating" we will create a new language.  For fun we are calling it "birrig" (for no reason).  It is a substitution cipher where we switch various letters for other letters, keeping the vowels and capitals and punctiation so it is somewhat readable.

Reading on chatGPT on a caesar cipher it is not able to consistently decode it after 3 letters.  We need to watch that GPT 4 doesn't fix this as it does better on many tasks but for now this should confuse it

So it should treat each word as an unknown and need to learn it as a new language.  In so doing we can simulate translating from English to birrig and then we have the cipher to see if it does the same.  

The weaknesses however are:
 - it is word for word translation and not requiring it to be reworded to choose best word
 - it is grammatically the same so it doesn't account for mechanics like changing word order for empasis in some langauges or cases like brother (older) is different from brother (younger).
